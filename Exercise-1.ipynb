# Solution-2
### Write a program to find out, if the given number is abundant…
#### Note: In number theory, an abundant number or excessive number is a number for which the sum of its proper divisors is greater than the number itself. The integer 12 is the first abundant number. Its proper divisors are 1, 2, 3, 4 and 6 for a total of 16.



```python
def sumOfProperDivisors(nu):
    
    candicateNumber = 1
    resultOfSum = 0
    
    while candicateNumber < nu:
        if nu % candicateNumber == 0:
            resultOfSum += candicateNumber
        candicateNumber += 1
    return resultOfSum

while True:
    s = input("\nLütfen Bir Sayı Giriniz: ")
    try:
        n = int(s)
    except:
        print("Sadece Sayı Giriniz")
        continue
    
    if sumOfProperDivisors(n) > n:
        print(f"Abundant Bir Sayı Buldunuz: {n}")
        break
    else:
        print(f"Maalesef {n} bir 'Abundant' sayı değil")
```

# Solution - 3
### Write a program that calculates the area and perimeter of selected 3 different geometric shapes…
#### Two values must be received from the user:
##### 1)Kind of the geometric shape(1:Circle, 2:Square, 3:Equilateral Triangle).
##### 2)The lengths of the sides or the radius of the circle.(If 2 or 3 are selected, the lengths of the sides. If 1 is selected, the radius of the circle.)
###### (Note: Assume the geometric shapes are regular, pi=3)


```python
def CircleWorks():
    import math
    
    cS =input("Please Give a Radius for Circle: ")
    cSn = float(cS)
    
    area = math.pi * cSn ** 2
    perimeter = 2 * math.pi * cSn
    
    print(f"""
    Circle Area: {area}
    Circle Perimeter: {perimeter}
    """)
    FinalQuestioner()
        
def SquareWorks():
    
    sS =input("Please Give a side lenght for Square: ")
    sSn = float(sS)
    
    area = sSn ** 2
    perimeter = sSn * 4
    
    print(f"""
    Square Area: {area}
    Square Perimeter: {perimeter}
    """)
    FinalQuestioner()

def EquilateralTriangleWorks():
    import math
    
    eS =input("Please Give a side lenght for Equilateral Triangle: ")
    eSn = float(eS)
    
    area = ((eSn ** 2) * math.sqrt(3))/4
    perimeter = eSn * 3
    
    print(f"""
    Square Area: {area}
    Square Perimeter: {perimeter}
    """)
    FinalQuestioner()

def Questioner():
    
    shapes = ["Circle","Square","Equilateral Triangle"]
    print(f"""\nShape Names

    1) {shapes[0]}
    2) {shapes[1]}
    3) {shapes[2]}
    """)
    
    s = input("\nPlease type a shape name or its index from Shape Names: ")
    if s.isdigit():
        ourDigit = int(s)
        if ourDigit == 1:
            CircleWorks()
        elif ourDigit == 2:
            SquareWorks()
        elif ourDigit == 3:
            EquilateralTriangleWorks()
        else:
            print("Hmmm Please a valid number")
            Questioner()
    else:
        if s == "Circle":
            CircleWorks()
        elif s == "Square":
            SquareWorks()
        elif s == "Equilateral Triangle":
            EquilateralTriangleWorks()
        else:
            print("Hmmm Please a valid name")
            Questioner()

def FinalQuestioner():
    while True:
        s = input("No-Tired? Type 'c' if not press enter: ")
        if s == "c":
            Questioner()
            break
        elif not s:
            print("Calculations Finished...")
            break
Questioner()
```
